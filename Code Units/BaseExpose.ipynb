```python
from web3 import Web3
from web3.exceptions import TransactionNotFound
```

```python
w3 = Web3(Web3.EthereumTesterProvider())
```

```python
w3.is_connected()
```

```python
#Provided Accounts and their balance
for account in w3.eth.accounts:
    balance = w3.eth.get_balance(account)
    print(f"Account: {account}, Balance: {Web3.from_wei(balance,'mether')}")
```

```python
#creating a new account and providing it 1 ether as initial balance
initial_balance = int(1e18)
govind = w3.eth.account.create()
```

```python
# Set initial balance for govind account
govind_transaction_hash = w3.eth.send_transaction({
        'from': w3.eth.accounts[0],
        'to': govind.address,
        'gas': 21000,
        'value': initial_balance
    })
```

```python
w3.eth.get_transaction(govind_transaction_hash)
```

```python
govind_transaction_receipt = w3.eth.wait_for_transaction_receipt(govind_transaction_hash)
```

```python
govind_transaction_receipt
```

```python
print("Address:", govind.address)
print("Key:", govind.key.hex)
print("Balance (Wei):", w3.eth.get_balance(govind.address))
print("Balance (Ether):", w3.from_wei(w3.eth.get_balance(govind.address), 'ether'))
```

```python
#Nonce: The number of transactions sent from this address.
print("Nonce:", w3.eth.get_transaction_count(govind.address))
```

```python
#Transaction Receipts: You can get information about the transactions involving this account.
try:
    transaction_receipt = govind_transaction_receipt
    if transaction_receipt:
        print("Transaction Receipts:", govind_transaction_receipt)
    else:
        print("Transaction Receipts: No transactions found for this account.")
except TransactionNotFound:
    print("Transaction Receipts: Transaction not found.")
```

```python
# Transaction History
print("Transaction History:")
latest_block = w3.eth.get_block('latest')
transactions = latest_block.transactions
for tx_hash in transactions:
    tx = w3.eth.get_transaction(tx_hash)
    print("Transaction Hash:", tx['hash'].hex())
    print("From:", tx['from'])
    print("To:", tx['to'])
    print("Value (Wei):", tx['value'])
    print("Value (Ether):", w3.from_wei(tx['value'], 'ether'))
    print("Gas Price (Wei):", tx['gasPrice'])
    print("Gas Price (Ether):", w3.from_wei(tx['gasPrice'], 'ether'))
    print("Gas Used:", tx['gas'])
    print("Gas Limit:", tx['gas'])
    print("Block Number:", tx['blockNumber'])
    print("Nonce:", tx['nonce'])
    print("Transaction Index:", tx['transactionIndex'])
    # Fetch block containing the transaction
    block = w3.eth.get_block(tx['blockNumber'])
    print("Block Timestamp:", block.timestamp)
    print("Block Hash:", block.hash.hex())
    print("Parent Hash:", block.parentHash.hex())

    # Fetch transaction receipt
    receipt = w3.eth.get_transaction_receipt(tx_hash)
    if receipt is not None:
        print("Transaction Receipt:")
        print("  Status:", "Success" if receipt['status'] == 1 else "Failure")
        print("  Cumulative Gas Used:", receipt['cumulativeGasUsed'])
        print("  Gas Used:", receipt['gasUsed'])

        # Print additional details for each log entry
        if 'logs' in receipt:
            print("  Logs:")
            for log in receipt['logs']:
                print("    Address:", log['address'])
                print("    Topics:", log['topics'])
                print("    Data:", log['data'])

        print("  Block Hash:", receipt['blockHash'].hex())
        print("  Block Number:", receipt['blockNumber'])
        print("  Transaction Index:", receipt['transactionIndex'])
        # Print contract address if transaction creates a contract
        if tx['to'] is None and 'contractAddress' in receipt:
            print("  Contract Address:", receipt['contractAddress'])
        # Print sender balance before transaction
        sender_balance_before = w3.eth.get_balance(tx['from'], block_identifier=tx['blockNumber'] - 1)
        print("  Sender Balance Before Transaction (Wei):", sender_balance_before)
        print("  Sender Balance Before Transaction (Ether):", w3.from_wei(sender_balance_before, 'ether'))
        # Print receiver balance before transaction if the transaction is a transfer
        if tx['to'] is not None:
            receiver_balance_before = w3.eth.get_balance(tx['to'], block_identifier=tx['blockNumber'] - 1)
            print("  Receiver Balance Before Transaction (Wei):", receiver_balance_before)
            print("  Receiver Balance Before Transaction (Ether):", w3.from_wei(receiver_balance_before, 'ether'))
        # Print sender balance after transaction
        sender_balance_after = w3.eth.get_balance(tx['from'], block_identifier='latest')
        print("  Sender Balance After Transaction (Wei):", sender_balance_after)
        print("  Sender Balance After Transaction (Ether):", w3.from_wei(sender_balance_after, 'ether'))
        # Print receiver balance after transaction if the transaction is a transfer
        if tx['to'] is not None:
            receiver_balance_after = w3.eth.get_balance(tx['to'], block_identifier='latest')
            print("  Receiver Balance After Transaction (Wei):", receiver_balance_after)
            print("  Receiver Balance After Transaction (Ether):", w3.from_wei(receiver_balance_after, 'ether'))
        # Print total transaction fee
        total_fee = tx['gasPrice'] * receipt['gasUsed']
        print("  Total Transaction Fee (Wei):", total_fee)
        print("  Total Transaction Fee (Ether):", w3.from_wei(total_fee, 'ether'))
        # Print transaction type description
        transaction_type = "Ether Transfer" if tx['to'] is not None else "Contract Deployment"
        print("  Transaction Type:", transaction_type)
        # Print effective gas price
        effective_gas_price = tx['value'] / tx['gas']
        print("  Effective Gas Price (Wei per Gas):", effective_gas_price)
        # Print nonce (Next)
        next_nonce = w3.eth.get_transaction_count(tx['from'], 'pending')
        print("  Nonce (Next):", next_nonce)
        # Print block confirmations
        block_confirmations = w3.eth.block_number - tx['blockNumber']
        print("  Block Confirmations:", block_confirmations)
        print("---------------------------")
```
